```python
import numpy as np
import pandas as pd
#import seaborn as sns
import matplotlib.pyplot as plt
```

```python
alb_mixed = pd.read_hdf("results/PuP/Alb_IS_Mixed/alb_IS_pup_mixed_merged.h5", "alb_IS_pup_mixed")
```

```python
alb_mixed
```

```python
total_trans_cost = alb_mixed["Trans_Cost"] + alb_mixed["Fixed_V"]
total_trans_cost.tolist()
```

```python
def funce(x):
    return ((x/118.36)*100)
```

```python
indexed_list = list(map(funce, total_trans_cost))
indexed_list = [round(num, 1) for num in indexed_list]
```

```python
indexed_list
```

```python
alb_mixed = alb_mixed.assign(ix_list = indexed_list)
```

```python
alb_mixed = alb_mixed.sort_values("ix_list")
alb_mixed
```

```python
def set_size(width_pt, fraction=1, subplots=(1, 1)):

    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)
```

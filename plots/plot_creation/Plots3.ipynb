```python
tex_fonts = {
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": ["Palatino"],
    "axes.labelsize": 12,
    "font.size": 12,
    "legend.fontsize": 10,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11
}

plt.rcParams.update(tex_fonts)
```

```python
def set_size(width_pt, fraction=1, subplots=(1, 1)):

    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)
```

```python
wue_IS_100_ = pd.read_hdf("results/Wue_IS100/wue_IS_100_merged.h5", "wue_IS_100")
wue_IS_50_ =  pd.read_hdf("results/Wue_IS50/wue_IS_50_merged.h5", "wue_IS50")
wue_IS_25_ =  pd.read_hdf("results/Wue_IS25/wue_IS_25_merged.h5", "wue_IS_25")
wue_IS_150_ = pd.read_hdf("results/Wue_IS150/wue_IS_150_merged.h5", "wue_IS150")
df_list_ = [wue_IS_25_, wue_IS_50_, wue_IS_100_, wue_IS_150_]
wue_IS = pd.concat(df_list_, ignore_index=True)
wue_IS
```

```python
wue_avg = wue_IS.groupby(np.arange(len(wue_IS))//10).mean()
wue_avg
```

```python
wue_sub = wue_avg[["FixedDC_Cost", "Trans_Cost", "Variable_DC", "Fixed_V"]]
#wue_sub["Customers"] = ["25C","50C","100C","150C"]
wue_sub
```

```python
wue_perc = wue_sub.divide(wue_sub.sum(axis=1), axis=0)
wue_perc = wue_perc * 100
```

```python
wue_perc["Customers"] = ["25C","50C","100C","150C"]
wue_perc
```

```python
xticks = [1,2,3,4]
labels = ["$25C$","$50C$", "$100C$", "$150C$"]
plt.style.use("seaborn-ticks")
fig, ax = plt.subplots(1,1, figsize=set_size(445))


ax.stackplot(wue_perc["Customers"],  wue_perc["Variable_DC"], wue_perc["Trans_Cost"],
              wue_perc["Fixed_V"], wue_perc["FixedDC_Cost"],labels = ["Variable DC","Transportation Cost", "Fixed Vehicle", "Fixed DC"])
ax.yaxis.set_major_formatter(plticker.PercentFormatter(100))                                                                      
                                             
plt.figlegend(loc="lower left", ncol=4)
plt.show()
fig.savefig("result_graph/stacked_area_test.png", format = "png")
```

```python
xticks = [1,2,3,4]
labels = ["$25C$","$50C$", "$100C$", "$150C$"]
plt.style.use("seaborn-ticks")
fig, ax = plt.subplots(1,1, figsize=set_size(1000))

customers = wue_perc["Customers"]
fixed_DC = wue_perc["FixedDC_Cost"]
var_DC = wue_perc["Variable_DC"]
trans = wue_perc["Trans_Cost"]
fixed_V=wue_perc["Fixed_V"]


# Plot lines
ax.plot(customers, fixed_DC, color="green")
ax.plot(customers, var_DC, color="red")

# Fill area when income > expenses with green
ax.fill_between(
    customers, fixed_DC, var_DC, where=(fixed_DC > var_DC), 
    interpolate=True, color="green", alpha=0.25, 
    label="fixed_DC"
)

# Fill area when income <= expenses with red
ax.fill_between(
    customers, fixed_DC, var_DC, where=(fixed_DC <= var_DC), 
    interpolate=True, color="red", alpha=0.25,
    label="var_DC"
)

ax.legend();
```

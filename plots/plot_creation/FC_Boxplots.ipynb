```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
tex_fonts = {
    "text.usetex": True,
    "font.family": "serif",
    "font.sans-serif": ["Palatino"],
    "axes.labelsize": 12,
    "font.size": 12,
    "legend.fontsize": 10,
    "xtick.labelsize": 11,
    "ytick.labelsize": 11
}

plt.rcParams.update(tex_fonts)
```

```python
def set_size(width_pt, fraction=1, subplots=(1, 1)):

    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)
```

```python
alb_F100 = pd.read_hdf("results/Alb_F100/alb_F_100_merged.h5", "alb_F_100")
alb_F50 =  pd.read_hdf("results/Alb_F50/alb_F_50_merged.h5", "alb_F_50")
alb_F25 =  pd.read_hdf("results/Alb_F25/alb_F_25_merged.h5", "alb_F_25")
alb_F150 = pd.read_hdf("results/Alb_F150/alb_F_150_merged.h5", "alb_F_150")
df_list_ = [alb_F25, alb_F50, alb_F100, alb_F150]
alb_F = pd.concat(df_list_, ignore_index=True)
alb_F["Stops_Distance"] = alb_F["Distance"] / alb_F["Stops"]
alb_F
```

```python
list_alb100 = alb_F["Total_Cost_per_order"].tolist()

list_alb100= [list_alb100[x:x+10] for x in range(0, len(list_alb100), 10)]
list_alb100
```

```python
wue_F100 = pd.read_hdf("results/Wue_F100/wue_F_100_merged.h5", "wue_F_100")
wue_F50 =  pd.read_hdf("results/Wue_F50/wue_F_50_merged.h5", "wue_F_50")
wue_F25 =  pd.read_hdf("results/Wue_F25/wue_F_25_merged.h5", "wue_F_25")
wue_F150 = pd.read_hdf("results/Wue_F150/wue_F_150_merged.h5", "wue_F_150")
df_list_ = [wue_F25, wue_F50, wue_F100, wue_F150]
wue_F = pd.concat(df_list_, ignore_index=True)
wue_F
```

```python
list_wue100 = wue_F["Total_Cost_per_order"].tolist()

list_wue100= [list_wue100[x:x+10] for x in range(0, len(list_wue100), 10)]
list_wue100
```

```python
plt.style.use("seaborn-ticks")
wue25 = list_wue100[0]
wue50 = list_wue100[1]
wue100 = list_wue100[2]
wue150 = list_wue100[3]
wue_demand = [wue25, wue50, wue100, wue150]

alb25 = list_alb100[0]
alb50 = list_alb100[1]
alb100 = list_alb100[2]
alb150 = list_alb100[3]
alb_demand = [alb25, alb50, alb100, alb150]

xticks = [1,2,3,4]
labels = ["$25C$","$50C$", "$100C$", "$150C$"]

fig, (ax1,ax2) = plt.subplots(1,2, figsize=set_size(445))

ax1.set_title("\\textbf{Albstadt}", fontsize=14)
ax2.set_title("\\textbf{Wrzburg}", fontsize=14)


ax1.set_ylabel("\\textbf{Cost/Order in }")

ax1.boxplot(alb_demand, 
           patch_artist=True, widths=0.65,
           boxprops=dict(facecolor="grey", color="black"),
           medianprops=dict(color="black"))
ax1.set_xticks(xticks)
ax1.set_xticklabels(labels, rotation = 0)

ax2.boxplot(wue_demand, 
           patch_artist=True, widths=0.65,
           boxprops=dict(facecolor="grey", color="black"),
           medianprops=dict(color="black"))

loc =plticker.MultipleLocator(base=3)
ax1.yaxis.set_major_locator(loc)
ax2.yaxis.set_major_locator(loc)

ax2.set_xticks(xticks)
ax2.set_xticklabels(labels, rotation = 0)

fig.supxlabel("\\textbf{Demand Scenario}", y = 0.05, fontsize=12)
ax1.yaxis.grid(True)
ax2.yaxis.grid(True)
#ax2.yaxis.grid(True)

fig.tight_layout()
plt.show()
fig.savefig("result_graph/FC_demand_scenarios.svg", format="svg")
```

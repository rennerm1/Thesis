```python
def set_size(width_pt, fraction=1, subplots=(1, 1)):

    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)
```

```python
wue_IS_100_ = pd.read_hdf("results/Wue_IS100/wue_IS_100_merged.h5", "wue_IS_100")
wue_IS_50_ =  pd.read_hdf("results/Wue_IS50/wue_IS_50_merged.h5", "wue_IS50")
wue_IS_25_ =  pd.read_hdf("results/Wue_IS25/wue_IS_25_merged.h5", "wue_IS_25")
wue_IS_150_ = pd.read_hdf("results/Wue_IS150/wue_IS_150_merged.h5", "wue_IS150")
df_list_ = [wue_IS_25_, wue_IS_50_, wue_IS_100_, wue_IS_150_]
wue_IS = pd.concat(df_list_, ignore_index=True)
wue_IS["Stops_Distance"] = wue_IS["Distance"] / wue_IS["Stops"]
wue_IS
```

```python
alb_IS_100_ = pd.read_hdf("results/Alb_IS100/alb_IS_100_merged.h5", "alb_IS_100")
alb_IS_50_ =  pd.read_hdf("results/Alb_IS50/alb_IS_50_merged.h5", "alb_IS_50")
alb_IS_25_ =  pd.read_hdf("results/Alb_IS25/alb_IS_25_merged.h5", "alb_IS_25")
alb_IS_150_ = pd.read_hdf("results/Alb_IS150/alb_IS_150_merged.h5", "alb_IS_150")
df_list_ = [alb_IS_25_, alb_IS_50_, alb_IS_100_, alb_IS_150_]
alb_IS = pd.concat(df_list_, ignore_index=True)
alb_IS["Stops_Distance"] = alb_IS["Distance"] / alb_IS["Stops"]
alb_IS
```

```python
wue_group_dist = wue_IS.groupby("Customers_N")["Stops_Distance"].mean().tolist()
wue_group_time = wue_IS.groupby("Customers_N")["Stops_h"].mean().tolist()
alb_group_dist = alb_IS.groupby("Customers_N")["Stops_Distance"].mean().tolist()
alb_group_time = alb_IS.groupby("Customers_N")["Stops_h"].mean().tolist()
```

```python
wue_dist_min = wue_IS.groupby("Customers_N")["Stops_Distance"].min().tolist()
wue__dist_max =wue_IS.groupby("Customers_N")["Stops_Distance"].max().tolist()
alb_dist_min = alb_IS.groupby("Customers_N")["Stops_Distance"].min().tolist()
alb_dist_max = alb_IS.groupby("Customers_N")["Stops_Distance"].max().tolist()

wue_time_min = wue_IS.groupby("Customers_N")["Stops_h"].min().tolist()
wue_time_max = wue_IS.groupby("Customers_N")["Stops_h"].max().tolist()
alb_time_min = alb_IS.groupby("Customers_N")["Stops_h"].min().tolist()
alb_time_max = alb_IS.groupby("Customers_N")["Stops_h"].max().tolist()
```

```python
alb_std_time = alb_IS.groupby("Customers_N")["Stops_h"].std().tolist()
alb_std_dist = alb_IS.groupby("Customers_N")["Stops_Distance"].std().tolist()

wue_std_dist = wue_IS.groupby("Customers_N")["Stops_Distance"].std().tolist()
wue_std_time = wue_IS.groupby("Customers_N")["Stops_h"].std().tolist()
```

```python
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color="#1f77b4", lw=2),
                Line2D([0], [0], color="#ff7f0e", lw=2)]


lb_alb_time = np.array(alb_group_time) - np.array(alb_std_time)
lb_alb_dist = np.array(alb_group_dist) - np.array(alb_std_dist)
ub_alb_time = np.array(alb_group_time) + np.array(alb_std_time)
ub_alb_dist = np.array(alb_group_dist) + np.array(alb_std_dist)

lb_wue_time =  np.array(wue_group_time) - np.array(wue_std_time)
lb_wue_dist = np.array(wue_group_dist) - np.array(wue_std_dist)
ub_wue_time = np.array(wue_group_time) + np.array(wue_std_time)
ub_wue_dist = np.array(wue_group_dist) + np.array(wue_std_dist)

base_x = ["$25C$","$50C$", "$100C$", "$150C$"]

plt.style.use("seaborn-ticks")

fig,(ax1,ax2) = plt.subplots(1,2,figsize=set_size(445))
ax3=ax1.twinx()
ax1.plot(base_x, alb_group_dist, marker = "o",ms=4, label="dist")
ax1.fill_between(base_x, lb_alb_dist, ub_alb_dist, alpha=0.15)

ax3.plot(base_x, alb_group_time, color = "#ff7f0e", marker = "x", ms=4, label="time")
ax3.fill_between(base_x, lb_alb_time, ub_alb_time, alpha = 0.15, color = "#ff7f0e")

ax4= ax2.twinx()
ax2.plot(base_x, wue_group_dist, marker = "o", ms=4)
ax2.fill_between(base_x, lb_wue_dist, ub_wue_dist, alpha = 0.3)
ax4.plot(base_x, wue_group_time, color="#ff7f0e", marker = "x", ms=4)
ax4.fill_between(base_x, lb_wue_time, ub_wue_time, alpha = 0.3, color="#ff7f0e")



ax1.set_ylabel("\\textbf{Meters/Stop}")
ax4.set_ylabel("\\textbf{Stops/h}")

fig.legend(custom_lines, ['Meters/Stop', 'Stops/h'], loc = "lower left", labelspacing=0.2,
          prop={"size":"small"})
fig.supxlabel("\\textbf{Demand Scenario}", y = 0.05, fontsize=12)
ax1.set_title("\\textbf{Albstadt}", fontsize=14)
ax2.set_title("\\textbf{Wrzburg}", fontsize=14)
plt.tight_layout()
plt.show()
#fig.savefig("result_graph/time_dist_std.svg", format ="svg")
```

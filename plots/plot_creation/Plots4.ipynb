```python
def set_size(width_pt, fraction=1, subplots=(1, 1)):

    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction
    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    return (fig_width_in, fig_height_in)
```

```python
from scipy.interpolate import interp1d
plt.style.use("seaborn-white")
fig, ax1 = plt.subplots(1,1, figsize=set_size(445))

x= np.array(xbase)
x1= np.array(xbase_short)
y1= np.array(D)
y2= np.array(FC)
y3 = np.array(IS)

x_new = np.linspace(x.min(), x.max(),1000)
x1_new = np.linspace(x1.min(), x1.max(), 1000)
f = interp1d(x, y1, kind='quadratic')
f2 = interp1d(x, y2, kind="quadratic")
f3 = interp1d(x1, y3, kind="quadratic")

y_D=f(x_new)
y_IS=f3(x1_new)
y_FC=f2(x_new)

ax1.plot(x1_new, y_IS, label= "In Store", color="#207720")
ax1.plot (x_new,y_D, label = "Darkstore", color="#234990")
ax1.plot(x_new, y_FC, label= "FC", color = "#E23726")


ax1.axes.yaxis.set_ticklabels([])
ax1.axes.xaxis.set_ticklabels([])
#ax1.yaxis.grid(True)
plt.xlabel("\\textbf{Number of Orders}")
plt.ylabel("\\textbf{EBIT}")
plt.axhline(y=0.65, color='grey', alpha=0.6, linestyle='-')
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='center right', frameon=True,
           fancybox=True, ncol=1, borderpad=0.25, labelspacing = 0.5,
           borderaxespad=0.1, bbox_to_anchor=([1, 0.6, -0.05, 0]), fontsize = 11)
fig.tight_layout()
plt.show()
fig.savefig("result_graph/ff_comp.svg", format="svg")
#plt.scatter (x, y)
```
